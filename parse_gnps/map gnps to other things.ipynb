```python
atlas_filename = '/Users/bpb/Downloads/atlas_df_pos_hilic_cs.csv'
atlas = nodes[['precursor mass','RTMean','INCHI','Smiles']].reset_index()
atlas['inchi_key'] = ''
idx = atlas['INCHI'] == 'N/A'
atlas.loc[idx,'INCHI'] = np.nan
idx = atlas['INCHI'] == ' '
atlas.loc[idx,'INCHI'] = np.nan
idx = atlas['Smiles'] == 'N/A'
atlas.loc[idx,'Smiles'] = np.nan
for i,row in atlas.iterrows():
    mol = None
    if not pd.isna(row['INCHI']):
        mol = Chem.MolFromInchi(str(row['INCHI']))
    if not pd.isna(row['Smiles']):
        smiles = str(row['Smiles']).strip()
        mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        ik = Chem.MolToInchiKey(mol)
        atlas.loc[i,'inchi_key'] = ik
atlas.rename(columns={'precursor mass':'mz','RTMean':'rt_peak','cluster index':'label'},inplace=True)
atlas['rt_peak'] = atlas['rt_peak']/60.0
atlas['rt_min'] = atlas['rt_peak'] - 0.1
atlas['rt_max'] = atlas['rt_peak'] + 0.1
atlas.drop(columns=['INCHI','Smiles'],inplace=True)
print(atlas.shape)
atlas.to_csv(atlas_filename)
atlas[atlas['inchi_key'] != ''].to_csv(atlas_filename.replace('.csv','_identified.csv'))
atlas.head()
```

# locate the indices of matching nodes

```python
def find_matching_nodes(target,df,cutoff=10):
    idx = abs(df['precursor mass'] - target) / df['precursor mass'] * 1.0e6 < cutoff
    return df[idx]

target_mzs = np.asarray([523.1249,167.070276,295.226])
for t in target_mzs:
    df_hits = find_matching_nodes(t,nodes)
    print(df_hits[['parent mass','precursor mass','subgraph size','subgraph index']])
    print('')

    
```
